# Confusion Matrix Implementation - Iris Dataset

1.Uses inbuilt logistic Regression and Iris Datasets
2.Prints the number of wrong predictions
3.Prints the Probabilites of Exactly how much each row belongs to a particular class
4.Prints the Confusion Matrix

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection
from sklearn.metrics import confusion_matrix
```

```python
#Loading Iris Dataset
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

```python
#Splitting data into train and Test
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
clf = LogisticRegression(multi_class ='auto' ,solver = 'liblinear')
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred_train = clf.predict(x_train)
y_pred_test = clf.predict(x_test)
```

```python
#Number of Wrong Predictions
print( (y_pred_train - y_train).sum() )
print( (y_pred_test - y_test).sum() )
```

```python
#Probability of classes
clf.predict_proba(iris.data)
```

```python
#Printing Confuion Matrix for Training data and Testing Data
print(confusion_matrix(y_pred_train, y_train))
print()
print(confusion_matrix(y_pred_test, y_test))
```
